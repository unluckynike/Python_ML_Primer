```python
import pandas as pd
```

```python
# 获取数据
order_products = pd.read_csv("./instacart/order_products__prior.csv")
products = pd.read_csv("./instacart/products.csv")
orders = pd.read_csv("./instacart/orders.csv")
aisles = pd.read_csv("./instacart/aisles.csv")
```

```python
products.head()
```

```python
orders.head()
```

```python
aisles.head()
```

```python
# 2、合并表
# order_products__prior.csv：订单与商品信息

# 字段：order_id, product_id, add_to_cart_order, reordered
# products.csv：商品信息
# 字段：product_id, product_name, aisle_id, department_id
# orders.csv：用户的订单信息
# 字段：order_id,user_id,eval_set,order_number,….
# aisles.csv：商品所属具体物品类别
# 字段： aisle_id, aisle

# 合并aisles和products aisle和product_id
table1=pd.merge(aisles,products,on=["aisle_id","aisle_id"])
```

```python
table1.head()
```

```python
table2=pd.merge(table1,order_products,on=["product_id","product_id"])
```

```python
table2.head()
```

```python
table3=pd.merge(table2,orders,on=["order_id","order_id"])
```

```python
table3.head() # 通过字段合并后的最终表
```

#### 透视表就是将指定原有DataFrame的列分别作为行索引和列索引，然后对指定的列应用聚集函数(默认情况下式mean函数)。交叉表是用于统计分组频率的特殊透视表

```python
# 3、找到user_id和aisle之间的关系 交叉表
table=pd.crosstab(table3["user_id"],table3["aisle"])
```

```python
table.head()
```

```python
data=table[:10000]# 取一万条数据
```

```python
from sklearn.decomposition import PCA
```

```python
transfer=PCA(n_components=0.95)
data_new=transfer.fit_transform(table)
```

```python
data_new.shape # 由原来的134个特征降到了44个特征
```
